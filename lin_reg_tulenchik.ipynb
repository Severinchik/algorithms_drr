```python
x = np.array([1, 2, 3, 4, 6, 8])
y = np.array([0.1, 0.7, 2, 2.4, 3, 3.1])
```

```python
A = np.vstack([x, np.ones(len(x))]).T
print (A)
```

```python
m, c = np.linalg.lstsq(A, y)[0]
print(m, c)
```

```python
plt.plot(x, y, 'o', label='os, shcho my maly', markersize=10)
plt.plot(x, m*x + c, 'r', label='peredbachennya')
plt.legend()
plt.show()
```
